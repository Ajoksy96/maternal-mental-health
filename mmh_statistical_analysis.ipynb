```python
# import relevant libraries & data
import pandas as pd
dataframe = pd.read_csv("data/cleaned_merged_adult_data.csv")
```

```python
df = dataframe.copy()
```

```python
# look at data
df
```

```python
# recode race variable to be binary - 0 white, 1 non-white
df["race_binary"] = df["race"]
for row in df['race_binary']: # recoded (5 = other)
    if row == 1:
        df['race_binary'] = df['race_binary'].replace(row, 0)
for row in df['race_binary']:
    if row > 1:
        df['race_binary'] = df['race_binary'].replace(row, 1)
```

```python
# look at data
df
```

```python
# anxiety frequency by race
white_anx = df[df['race_binary'] == 0.0]['anx_frequency']
nonwhite_anx = df[df['race_binary'] == 1.0]['anx_frequency']
```

```python
# depression frequency by race
white_dep = df[df['race_binary'] == 0.0]['dep_frequency']
nonwhite_dep = df[df['race_binary'] == 1.0]['dep_frequency']
```

```python
# anxiety frequency by year
anx_2019 = df[df['survey_year'] == 2019]['anx_frequency']
anx_2020 = df[df['survey_year'] == 2020]['anx_frequency']
anx_2021 = df[df['survey_year'] == 2021]['anx_frequency']
```

```python
# depression frequency by year
dep_2019 = df[df['survey_year'] == 2019]['dep_frequency']
dep_2020 = df[df['survey_year'] == 2020]['dep_frequency']
dep_2021 = df[df['survey_year'] == 2021]['dep_frequency']
```

```python
# t-test to see racial differences (white vs. non-white) in average anxiety
import pingouin
ttest_race_anx = pingouin.ttest(white_anx, nonwhite_anx)
ttest_race_anx
```

```python
# average anxiety frequency by race
avg_anx_race = df.groupby(["race_binary"])["anx_frequency"].mean()
avg_anx_race
```

```python
# t-test to see racial differences (white vs. non-white) in average depression
ttest_race_dep = pingouin.ttest(white_dep, nonwhite_dep)
ttest_race_dep
```

```python
# average depression frequency by race
avg_dep_race = df.groupby(["race_binary"])["dep_frequency"].mean()
avg_dep_race
```

```python
# t-test to see yearly differences in average anxiety
ttest_year_anx_1 = pingouin.ttest(anx_2019, anx_2020)
ttest_year_anx_1
```

```python
# average anxiety frequency by year
avg_anx_race = df.groupby(["survey_year"])["anx_frequency"].mean()
avg_anx_race
```

```python
# t-test to see yearly differences in average anxiety
ttest_year_anx_2 = pingouin.ttest(anx_2019, anx_2021)
ttest_year_anx_2
```

```python
# recode relevant binary variables to be 0 absence, 1 presence
for row in df['wic_benefits_12m']: # recoded (0 no, 1 yes)
    if row == 2:
        df['wic_benefits_12m'] = df['wic_benefits_12m'].replace(row, 0)
for row in df['us_born']: # recoded (0 no, 1 yes)
    if row == 2:
        df['us_born'] = df['us_born'].replace(row, 0)
for row in df['health_insur']: # recoded (0 no, 1 yes)
    if row == 2:
        df['health_insur'] = df['health_insur'].replace(row, 0)
for row in df['med_care_12m']: # recoded (0 no, 1 yes)
    if row == 2:
        df['med_care_12m'] = df['med_care_12m'].replace(row, 0)
for row in df['pregnant_now']: # recoded (0 no, 1 yes)
    if row == 2:
        df['pregnant_now'] = df['pregnant_now'].replace(row, 0)
        
for row in df['marital_stat']: # recoded (0 single, 1 married)
    if row == 2:
        df['marital_stat'] = df['marital_stat'].replace(row, 0)
for row in df['sex_orient']: # recoded (0 straight, 1 gay)
    if row == 2:
        df['sex_orient'] = df['sex_orient'].replace(row, 0)
```

```python
# multiple linear regression with anxiety (model 1)
import statsmodels.formula.api as sm
formula1 = 'anx_frequency ~ survey_year + wic_benefits_12m + age + inc_pov_ratio + phys_health + bmi + region + education + race_binary'
model1 = sm.ols(formula = formula1, data = df)
reg1 = model1.fit()
print(reg1.summary())
```
